```python
#Answer 3
"""
Dispersion refers to the spread or variablity of a dataset. It measures how far individual data points are from central tendency
and helps describe the distribution's cobsistency, variability, or diversity.

Variance:
Variance measures the average squraed deviation of each data point from the mean.
It gives a sense of how spread out the data is.
A higher value of variance indicates high spread in the data set.
Ex:A test score with high variance value indicates that student performance varies widely.

Standard Deviation:
The standard deviation is the square root of the variance.
It measures the average deviation of data points from the mean in the same units as the original data.
A low standard deviation indicates that data points are close to the mean, signifying consistency.
A high standard deviation indicates more spread, implying variability.
"""
```

```python
#Answer 7
"""
The interquartile range is a measure of the spread of the middle 50% of a dataset.
It is calculated as the difference between third qurtile(Q3) and first qurtile(Q1).
                              IQR = Q3 - Q1

Outliers are data points that fall significantly outside the typical range of the dataset.
The 1.5*IQR rule is commonly used to identify outliers:
1.Calculate the IQR: Q3 - Q1
2.Determine the Lower bound: Q1 -(1.5*IQR)
3.Determine the upper bound: Q3 +(1.5*IQR)
4.Identify the outliers: Any point less than lower bound or greater than upper bound is considered outliers.
"""

```
