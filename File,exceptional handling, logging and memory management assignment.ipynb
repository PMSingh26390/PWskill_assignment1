```python
#Answer 2
"""
Exception handeling in python is a mechanism to handle errors that occurs during program execution in a controlled way,
preventing the program from crashing and allowing us to address specific error conditions.
"""
try:
  num = int(input("enter a number"))
  result = 100 / num
  print (f"result {result}")
except ZeroDivisionError:
  print("can not divide by zero")
except ValueError:
  print("invalid input, enter a valid number")
else:
  print("no error occured")
finally:
  print("execution compeleted")
```

```python
#Answer 24
"""
We can manually raise an exception in python by using the 'raise' statement.
This allows us to trigger an execption under specific conditions
"""
class ManualError(Exception):
  def __init__(self,msg):
    super().__init__(msg)

raise ManualError("this is a manual error")
```

```python
#Answer 7
import logging
logging.basicConfig(filename="diff_level.log", level = logging.INFO, format= '%(asctime)s %(level.name)s %(message)s')
try:
  x = int(input("enter the numberator"))
  y = int(input("enter the denominator"))
  result = x / y
  logging.info("result %s",result)

except (ZeroDivisionError,ValueError) as e:
  if y == 0:
    logging.error("can not divide by zero: %s",e)
  elif y != int:
    logging.warning("denominator should be an integer")

else:
  logging.info("no error occured")

finally:
  logging.info("execution compeleted")


```
