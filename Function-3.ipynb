```python
#Answer 1
"""
Function = It is block of reusable code that is not tied to any specific objects.
Function are typically at the module level or standalone.
"""
#Example
def add(a,b):
  return a+b
print(add(2,3))
"""
Method = It is basically a function which associated with an object. Methods are defined inside a class
and operate on the data specific to that instance of the class.
"""
#Example
class Greeter:
  def greet(self,name):
    return f"hello,{name}"

g = Greeter()
print(g.greet("world"))
```

```python
# Answer 2
"""
Function Parameters = They are the variable names defined in the function's declaration. they act as placeholder
for the values thst the function will receive when it is called.

Function Arguments = They are the actual values or data passed to the function when it is called. These values are assigned
to the function's parameters
"""
#Example
def add(a,b): # a,b are function parameters
  return a+b
print(add(2,3)) # 2,3, are function arguments
```

```python
# Answer 3
"""
1. Standard Function Definations: A function is defined using "def" keyword, followed by a name,
parentheses (for parameter), and a block of code
"""
#Example
def add(a,b): #"add" name of the function, "a,b" are function parameter
  return a+b # block of code
print (add(2,3)) # Calling a function


"""
2. Lambda Function: A one line function defined using the lambda keyword, usually for short,
simple tasks.
"""
# Example
add = lambda a,b:a+b # "add" name of fuction,"a,b" functions parameter, "a+b" operation to be carried on parameters
print(add(2,3)) # calling a lambda function


"""
3. Recursive Function: A function that calls itself to solve a smaller sub problem.
"""
# Eample
def facto(n):# defining a fuction
  if n==0:
    return 1
  else:
    return n*facto(n-1) # calling the defined fuction
print(facto(5))
```

```python
# Answer 5
"""
An iterable is any object that can return an iterator, meaning itimplements the special method
"iter()".
We can loop over an iterable.
It does not keep track of its current state, we must get an iterator to traverse it.
"""
#Example
l = [1,2,3,4,5] # List is an iterable
for i in l:
  print(i)

"""
An Iterator ia an object that represent a stream of data. It implements the special methods "iter()" and "next()"
Keeps track of its current state.
Produces the next value in the sequence when we use "next()".
once all items are consumed, its raises a stopiteration exception.
"""
#Example
l = [1,2,3,4,5] # List is an iterable
iter = iter(l)
print(next(iter)) # output = 1
print(next(iter)) # output = 2
print(next(iter)) # output = 3
print(next(iter)) # output = 4
print(next(iter)) # output = 5
```

```python
# Answer 6
"""
A generator is a special type of function that produces a sequence of value one at a time instead of computing them
all at once and storing them in memory. Generators are used to handle large data sets or infinite sequences.

Generators are defined like regular functions but use the "yield" keyword instead of "return" to produce values.
"""
#Example
def square(n):# define a fucntion
  for i in range(n):
    yield i**2 # will give one value at a time
for i in square(5): # using loop statement to print all output
  print(i)
```

```python
# Answer 8
"""
A lambda function in python is a small, anonymous function that is defined using
the lambda keyword.
Lambda functions are single-expression functions that can take any number of arguments but can only have one expression.
Lambda function are generally used when:
1. Passing as an argument to a higher order function
2. In data processing pipeline
3. Defining simple one liner
"""
#Example
add = lambda a,b:a+b
print(add(2,3))
```

```python
# Answer 9
"""
The map() function in python is used to apply a specified function to each item of an iterable
and return a map object containing the result.It is a poerful tool for transforming data efficiently and concisely.
The primary purpose of map() is to avoid explicit loops when performing operations on very item in an iterable.
it makes code cleaner and often more efficient.
"""
#Example: Multiply each element of a list by 2
l = [1,2,3,4,5]
result = map(lambda x:x*2,l)
print(list(result))
```

```python
# Answer 10
"""
 "map()" transforms each elements in the iterable using the function.
"""
# #Example: Square each element of a list by 2
l = [1,2,3,4,5]
result = map(lambda x:x**2,l)
print(list(result))

"""
"reduce()" aggregates all elements into a single value by applying the function cumulatively.
"""
#Example: Calculate the product of all elements in a list
from functools import reduce
l = [1,2,3,4,5]
result = reduce(lambda x,y:x*y,l)
print(result)

"""
"filter()" filters elements from an iterable based on a condition specified by the function.
"""
#Example: Finding even number in a list
l = [1,2,3,4,5]
result = filter(lambda x:x%2==0,l)
print(list(result))
```

```python
# Answer 6
def expo_pow(n):
  for i in range (n):
    x = 2**i
    yield x
expo_pow(5)
iter = expo_pow(5)
next(iter)
```

```python
# Answer 7
def rw(n):
  line = ""
  for i in n:
    if i == "\n":
      yield line
      line = ""
    else:
      line = line + i
  yield line
rw("hhello world \n welcome to python")
iter = rw("hello world \n welcome to python")
next(iter)
```

```python
# Answer 10
# Method 1
def novowels(n):
  vo = "aeiouAEIOU"
  for i in n:
    if i in vo:
      continue
    else:
      return i

print("".join(filter(novowels,"hello world")))

# Method 2
def novowels(n):
  vo = "aeiouAEIOU"
  return n not in vo

print("".join(filter(novowels,"this an example")))
```

```python
# Answer 11
# Method 1
ord =[[34587,40.95,4],[98762,56.80,5],[77226,32.95,3],[88112,24.99,3]]
def cal_ord(ord):
  ord_number,price_per_item,quantity = ord
  total = price_per_item*quantity
  if total < 100:
    quantity += 10
  return ord_number,quantity
new_ord = (list(map(cal_ord,ord)))
print(f"new order {new_ord}")


# Method 2
ord =[[34587,40.95,4],[98762,56.80,5],[77226,32.95,3],[88112,24.99,3]]
result = list(map(lambda x:(x[0],x[2]+10 if x[2]*x[1]<100 else x[2]),ord))
print(result)

```
