```python
# Answer 12
"""
Method overriding in OOPs occurs when a subclass provides a specific implementation of a method that is already defined in its superclass.
The method in the subclasshas the same name , parameters and return type as same in the super class, but itvoverrides the functionality
with its own version.
the key concept is that the subclass overrides the behaviour of the inherited method, allowing it to provide speciallized behaviour while
retaining the method's signature.
"""
class Animal:
  def sound(self):
    print("animals make sounds")
class Dog(Animal):
  def sound(self):
    print("Dog barks")
class Cat(Animal):
  def sound(self):
    print("cat meows")
# Object creation
obj = Animal()
dog = Dog()
cat = Cat()
#Calling overridden method
obj.sound()
dog.sound()
cat.sound()
```

```python
#Answer 13
"""
The "property decorator" in python is used to define a method as a property, allowing you to access it like an attribute rather than
a method.
It is often used to create getter, setter and deleter methods.
"""
class Circle:
  def __init__(self,radius):
    self.radius = radius

  @property
  def circle_radius(self):
    return self.radius

  @circle_radius.setter
  def radius_setter(self,value):
    if value < 0:
      raise ValueError("Radius cannot be negative")
      self.radius = value

  @property
  def area(self):
    return 3.14 * self.radius **2


obj = Circle(10)
print(obj.circle_radius)
print(obj.area)
obj.radius = 20
print(obj.circle_radius)
print(obj.area)
```

```python
#Answer 15
"""
An abstract class in python is a class that cannot be instantiated directly and is designed to be a blueprint for other classes.
It is defined using the "abc" module which is abstract base classes and typically contains one or more abstract methods
"""
from abc import ABC, abstractmethod

class Animal(ABC):
  @abstractmethod
  def sound(self):
    pass

class Dog(Animal):
  def sound(self):
    print("Dog barks")

class Cat(Animal):
  def sound(self):
    print("Cat meows")


dog = Dog()
cat = Cat()
dog.sound()
cat.sound()
```

```python
#Answer 23
"""
Polymorphism in python, especially with inheritance, allows object of different classes to be treated as objects of a common base
class.
It enables the same interface to be used for different underlying data types or classes.
Polymorphism is achieved through method overriding and duck typing.
"""

#Example
class Animal:
  def sound(self):
    print("Animals make sounds")

class Dog(Animal):
  def sound(self):
    print("Dog barks")

class Cat(Animal):
  def sound(self):
    print("Cat meows")

# Polymorphic Behavior
def make_sound(animal):
  animal.sound()

dog = Dog()
cat = Cat()
make_sound(dog)
make_sound(cat)
```

```python
#Answer 24
"""
Method chaining in python OOPs is a programming technique where multiple methods are called on the same object
sequentially in a single line.
Each method returns the object itself, allowing us to chain method calls together.
"""
#Example
class Person:
  def __init__(self,name):
    self.name = name
    self.age = None
    self.city = None

  def set_age(self,age):
    self.age = age
    return self

  def set_city(self,city):
    self.city = city
    return self

  def show(self):
    print(f"Name: {self.name}, Age: {self.age}, City: {self.city}")

p = Person("Tom")
p.set_age(20).set_city("Ranchi").show()

```

```python
#Answer 25
"""
The "__call__" method in python is a special or magic method that allows an instance of class to be called as
if it were a function.
When we define the "__call__"method in a class, we can make the object of the class callable, meaning we can use
the object like a function, passing arrguments and getting return values.
"""
#Example
class Adder:
  def __init__(self,a):
    self.a = a

  def __call__(self,b):
    return self.a + b

a = Adder(10)
print(a(34))
```

```python
#Answer 2

from abc import ABC, abstractclassmethod
class Shape(ABC):
  @abstractclassmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self,radius):
    self.radius = radius

  def area(self):
      return 3.14 * self.radius ** 2

class Rectangle(Shape):
  def __init__(self,lenght,width):
    self.lenght = lenght
    self.width = width

  def area(self):
    return self.lenght * self.width

circle = Circle(10)
rectangle = Rectangle(20,10)
print(circle.area())
print(rectangle.area())
```

```python
#Answer 12
class Student:
  def __init__(self,name,grade):
    self.name = name
    self.grade = grade

  def average_grade(self):
    if 91 <= self.grade <= 100:
      print("Grade A")
    elif 81 <= self.grade <= 91:
      print("Grade B")
    elif 71 <= self.grade <= 81:
      print("Grade C")
    elif 61 <= self.grade <= 71:
      print("Grade D")
    elif 51 <= self.grade <= 61:
      print("Grade E")
    else:
      print("Grade F")

student = Student("Priyam",98)
student.average_grade()
```

```python
#Answer 13
class Rectangle:

  def __init__(self):
    self.lenght = 0
    self.width = 0

  def set_dimensions(self,lenght,width):
    if lenght <= 0 or width <= 0:
      raise ValueError("Length and width must be positive values")
    self.lenght = lenght
    self.width = width

  def area(self):
    return self.lenght * self.width

obj = Rectangle()
obj.set_dimensions(34,56)
print(obj.area())
```

```python
#Answer 14
class Employee:
  def __init__(self):
    self.hours_worked = 0
    self.hourly_rate = 0
    self.salary = 0

  def calculate_salary(self,hours_worked,hourly_rate):
    if hours_worked <= 0 or hourly_rate <= 0:
      raise ValueError("Enter a positive value")
    self.hours_worked = hours_worked
    self.hourly_rate = hourly_rate
    self.salary = hours_worked * hourly_rate

  def bonus(self):
    if self.salary >= 50000:
      self.salary = self.salary + 0.1 * self.salary


emp = Employee()
emp.calculate_salary(160,507)
print(emp.salary)
emp.bonus()
print(emp.salary)

emp = Employee()
emp.calculate_salary(90,507)
print(emp.salary)
emp.bonus()
print(emp.salary)
```

```python
#Answer 18
class House:
  def __init__(self,address,price):
    self.address = address
    self.price = price

  def __str__(self):
    return f"House at {self.address} costs {self.price}"

class Mansion(House):
  def __init__(self,address,price,number_of_rooms):
    super().__init__(address,price)
    self.number_of_rooms = number_of_rooms

    def __str__(self):
      return f"Mansion at {self.address} costs {self.price} and has {self.number_of_rooms} rooms"

house = House("Kanke Road", 5000000)
print(house)
mansion = Mansion("Kanke Road", 50000000, 10)
print(mansion)
```
