```python
# Answer 2
'''
Mutable data are objects which can be modified after they are created, any modification occurs on same memory location.
List, dictionary, and set are examples of mutable data.
'''
# Example :
List1 = [1,2,3] # creating a list
List1[2] = 4 # changing data at memory loction 2
print(List1)# output = [1,2,4]


'''
Immutable data are objects which cannot be modified after they are created,any modification occurs on different memory location.
Tuple and string are examples of immutable data.
'''

# Example :
Tuple1 = (1,2,3) # creating a tuple
Tuple1[2] = 4 # changing data at memory loction 2
print(Tuple1)# output = TypeError: 'tuple' object does not support item assignment
```

```python
# Answer 5
'''
We might use set over list in python for following resons:

1. Uniqueness of elements: Duplicate elements are automatically removed in sets whereas duplicate elements are
    manually handled in list.

2. Faster Membership Testing: Sets are implemented using a hash table, making membership tests very efficient whereas lists require a linear
    search for membership testing.

3. Operations on Sets: Sets support various operations such as union, intersection, difference, and symmetric difference, which are not
    directly available in lists.

4. Better For Large Collections: Sets use less memory for large collections compared to lists.

5. Order Irrelevance: Sets do not maintain the order of elements.
'''

```
