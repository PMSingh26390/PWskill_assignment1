```python
#Answer 9
"""
Z Score for population with known variance is calculated using formula:
    Z(Score)= (X-MU)/SIGMA
    Where,
    X=Observed Value
    MU=Mean Of Population
    SIGMA=Standard Deviation Of Population


Z Score for sample with known value of variance is calculated using formula:
    Z(Score)= (X-MU)/SIGMA/SQRT(N)
    Where,
    X=Observed Value
    MU=Mean Of Population
    SIGMA=Standard Deviation Of Population
    N=Sample Size


If Z Score is higher than critical value, then reject null hypothesis
If Z Score is lower than Critical value, then fail to reject null hypothesis
"""
```

```python
#Answer 9
import numpy as np
import scipy.stats as stats
def confidence_interval(data,confidence_level = 0.95):
  n = len(data)
  sample_mean = np.mean(data)
  std_err = stats.sem(data)
  t_critical = stats.t.ppf((1 + confidence_level)/2,df = n-1)
  margin_of_error = t_critical * std_err
  lower_bound = sample_mean - margin_of_error
  upper_bound = sample_mean + margin_of_error
  return lower_bound,upper_bound

sample_data = [23,25,27,30,22,24,26,28,21,29]
lower,upper = confidence_interval(sample_data)
print(f"95% Confidence Interval:({lower:.2f},{upper:.2f})")
```

```python
#Answer 16
import numpy as np
import scipy.stats as stats
def f_test(data_1,data_2,alpha = 0.05):
  var_1 = np.var(data_1,ddof=1)
  var_2 = np.var(data_2,ddof=1)
  if var_1 > var_2:
    F = var_1/var_2
    dfn,dfd = len(data_1) - 1,len(data_2) - 1
  else:
    F = var_2/var_1
    dfn,dfd = len(data_2) - 1,len(data_1) - 1
  p_value = 2 * min(stats.f.cdf(F,dfn,dfd),1-stats.f.cdf(F,dfn,dfd))
  print(f"F-Statistic:{F:.4f}")
  print(f"p-value:{p_value:.4f}")
  if p_value < alpha:
    print("Reject the null hypothesis")
  else:
    print("Fail to reject the null hypothesis")


np.random.seed(42)
data_1 = np.random.normal(loc=0,scale=1,size=100)
data_2 = np.random.normal(loc=0,scale=1.2,size=100)
f_test(data_1,data_2)
```

```python
#Answer 19
import numpy as np
import scipy.stats as stats
def check_anova_assumptions(*groups,alpha=0.05):
  for i, group in enumerate(groups,start=1):
    stat,p_value = stats.shapiro(groups)
    normal = p_value > alpha
    if not normal:
      print(f"Group {i} is not normally distributed")
    else:
      print(f"Group {i} is normally distributed")


    stat,p_value = stats.levene(*groups)
    variance = p_value > alpha
    if not variance:
      print(f"Group {i} has equal variances")
    else:
      print(f"Group {i} has unequal variances")


np.random.seed(42)
group1 = np.random.normal(loc=50,scale=5,size=50)
group2 = np.random.normal(loc=55,scale=5,size=50)
group3 = np.random.normal(loc=60,scale=5,size=50)
check_anova_assumptions(group1,group2,group3)

```

```python
#Answer 20
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

np.random.seed(42)

diet_types = ['low-carb','Mediterranean','Vegan']
exercise_level = ['Low','Medium','High']

data = []
for diet in diet_types:
  for exercise in exercise_level:
    weight_loss = np.random.normal(loc=5 + 2*diet_types.index(diet) + exercise_level.index(exercise),scale=1.5,size=10)
    for wl in weight_loss:
      data.append([diet,exercise,wl])

df = pd.DataFrame(data,columns=['Diet','Exercise','Weight_Loss'])

model = ols('Weight_Loss ~ C(Diet) + C(Exercise) + C(Diet):C(Exercise)',data=df).fit()
anova_table = sm.stats.anova_lm(model,typ=2)
print(anova_table)

plt.figure(figsize=(10,6))
sns.boxplot(x='Diet',y='Weight_Loss',hue='Exercise',data=df,palette='Set2')
plt.xlabel('Diet')
plt.ylabel('Weight Loss')
plt.title('Weight Loss by Diet and Exercise')
plt.legend(title='Exercise')
plt.show()

```

```python
#Answer 24
import numpy as np
import scipy.stats as stats
def chi2_variance_test(sample,pop_var,alpha = 0.05):
  sample_var = np.var(sample,ddof=1)
  n = len(sample)
  chi2_stats = (n-1) * sample_var/pop_var
  p_value =2 * min(stats.chi2.cdf(chi2_stats,df=n-1),1-stats.chi2.cdf(chi2_stats,df=n-1))
  reject_null = p_value < alpha
  interprettation = "Reject the null hypothesis" if reject_null else "Fail to reject the null hypothesis"
  return chi2_stats,p_value,interprettation

np.random.seed(42)
sample = np.random.normal(loc=100,scale=10,size=30)
pop_var = 100
chi2_stats,p_value,interpretation = chi2_variance_test(sample,pop_var)
print(f"Chi-Square Statistic:{chi2_stats:.4f}")
print(f"p-value:{p_value:.4f}")
print(interpretation)
```

```python
#Answer 25
import numpy as np
import scipy.stats as stats

def z_test_proportions(success1,n1,success2,n2,alpha = 0.05):
  p1 = success1/n1
  p2 = success2/n2

  p_pool = (success1 + success2)/(n1 + n2)
  std_err = np.sqrt(p_pool * (1-p_pool) * (1/n1 + 1/n2))

  z_stats = (p1 - p2)/std_err

  p_value = 2 * (1 - stats.norm.cdf(np.abs(z_stats)))

  reject_null = p_value < alpha

  interpretation = "Reject the null hypothesis" if reject_null else "Fail to reject the null hypothesis"

  return z_stats,p_value,interpretation

success1 = 45
n1 = 100
success2 = 50
n2 = 100
z_stats,p_value,interpretation = z_test_proportions(success1,n1,success2,n2)

print(f"Z-Statistic:{z_stats:.4f}")
print(f"p-value:{p_value:.4f}")
print(interpretation)
```

```python
#Answer 26
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def f_test(data_1,data_2,alpha = 0.05):
  var_1 = np.var(data_1,ddof=1)
  var_2 = np.var(data_2,ddof=1)
  if var_1 > var_2:
    F = var_1/var_2
    dfn,dfd = len(data_1) - 1,len(data_2) - 1
  else:
    F = var_2/var_1
    dfn,dfd = len(data_2) - 1,len(data_1) - 1
  p_value = 2 * min(stats.f.cdf(F,dfn,dfd),1-stats.f.cdf(F,dfn,dfd))
  print(f"F-Statistic:{F:.4f}")
  print(f"p-value:{p_value:.4f}")
  if p_value < alpha:
    print("Reject the null hypothesis")
  else:
    print("Fail to reject the null hypothesis")


np.random.seed(42)
data_1 = np.random.normal(loc=0,scale=1,size=100)
data_2 = np.random.normal(loc=0,scale=1.2,size=100)
f_test(data_1,data_2)

sns.histplot(data_1,color='red',label="Sample1(Variance:{:.2f})".format(np.var(data_1,ddof=1)),kde=True,alpha=0.6)
sns.histplot(data_2,color='blue',label="Sample2(Variance:{:.2f})".format(np.var(data_2,ddof=1)),kde=True,alpha=0.6)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

```

```python
#Answer 27
import numpy as np
import scipy.stats as stats

def goodness_of_fit(observed,expected):
  chi2_stats,p_value = stats.chisquare(f_obs=observed,f_exp=expected)
  return chi2_stats,p_value

#Rolling a dice 60 times [1,2,3,4,5,6]
observed = np.array([8,12,10,15,9,6])
expected = np.full(len(observed),np.mean(observed))
chi2_stats,p_value = goodness_of_fit(observed,expected)
print(f"chi2_stats:{chi2_stats:.4f}")
print(f"p_value:{p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("the observed data does not follow the expected distribution")
else:
    print("the observed data follows the expected distribution")
```
